```python
import matplotlib.pyplot as plt
def plot_box(data,var):
    data.boxplot([var])
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt
def plot_box(data,var):
    data.boxplot(column=[var])
    plt.grid(False) #desactive la maya
    plt.show()
    
```

```python
for var in v_numericas:
    plot_box(data,var)
```

```python
# Rango Intercuartilico
def rango_inter(data,var):
    Q1=data[var].quantile(0.25)
    Q3=data[var].quantile(0.75)
    IQR=Q3-Q1
    lim_inf=Q1-3*IQR
    lim_sup=Q3+3*IQR
    lista=data.index[(data[var]<lim_inf)|(data[var]>lim_sup)]
    return lista
```

```python
lista_indices=[]
for var in v_numericas:
    lista_indices.extend(rango_inter(data,var))
lista_indices
```

### -----------------------------Eliminando atípicos--------------

```python
def remover(data,lista):
    lista=sorted(set(lista))# sorted es ordenar
    data=data.drop(lista)
    return data
data_limpia=remover(data,lista_indices)
data_limpia
```

```python
for var in v_numericas:
    plot_box(data_limpia,var)
```

### MÉTODO DE PUNTUACIÓN Z_elimin atipicos

#### Para usar esto tenemos que ver antes si tiene una distribución normal

```python
from sklearn import preprocessing
def puntuacion_z(data,var):
    z=preprocessing.scale(data[var])
    lista=data.index[abs(z)>2]
    return lista
```

```python
lista_indices=[]
for var in v_numericas:
    lista_indices.extend(puntuacion_z(data,var))
len(lista_indices)
```

### DISTANCIAS DE MAHALANOBIS

```python
#### Se hace un registro de outliers multivariados
```

```python
from scipy.spatial.distance import mahalanobis
import scipy as sp
X=data[v_numericas]
Sx=X.cov().values
IC=sp.linalg.inv(Sx)
media=X.mean().values
```

```python
def mahalanobisR(X,media,IC):
    m=[]
    for i in range(X.shape[0]):
        m.append(mahalanobis(X.iloc[i,:],media,IC)**2)
    return(m)

data_limpia['distMaha']=pd.DataFrame(mahalanobisR(X,media,IC),columns=['distMaha'])
data_limpia

data_limpia['index']=data_limpia.index
data_limpia
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_limpia['index'],data_limpia['distMaha'])
```

```python
data_elim=data_limpia[data_limpia['distMaha']>6]
data_elim
```

```python
elim=data_elim['index']
data=data_limpia.drop(elim,axis=0)
data=data.drop(['distMaha','index'],axis=1)
data
```
